## Exercise 1: Deterministic Environment (0 Marks)

In this exercise, the agent is able to move in the intended direction with certainty. For example, if it executes action _north_ in grid square 0, it will transition to grid square 5 with probability 1. In other words, we have a deterministic environment.

Compute the optimal policy using Value Iteration. 

Your need to produce two one-dimensional numpy arrays with names `policy` and `v`. Both arrays should have a length of 25, with the element at index $i$ representing grid cell $i$ (see figure above). Both arrays should be accessible in the "solution cell" below!

The array `policy` should be a numpy array of strings that specifies an optimal action at each grid location. Please use the abbreviations `"n"`, `"e"`, `"s"`, and `"w"` for the four actions. As an example, the value of `policy` at index `0` needs to give `"n"`, if _north_ is an optimal action in cell 0. The policy for a terminal state can be any action. If there are multiple optimal actions from a state, any optimal action will be considered as a correct answer. 

The array `v` should be an array of floats that contains the expected return at each grid square (that is, the state value under the optimal policy).

```python
import numpy as np

class Gridworld:
    def __init__(self):
        
        # Initializing environmental variables to be used
        self.num_rows = 5
        self.num_cols = 5
        self.num_fields = self.num_cols * self.num_rows
        self.gold_reward = 10
        self.bomb_reward = -10
        self.gold_positions = np.array([23])
        self.bomb_positions = np.array([18])

        self.rewards = np.zeros(shape=self.num_fields)
        self.rewards[self.bomb_positions] = self.bomb_reward
        self.rewards[self.gold_positions] = self.gold_reward

        self.actions = ["n", "e", "s", "w"]
        self.num_actions = len(self.actions)

        
    def make_step(self, state, curr_action, model):
        
        
        # Choosing model for the different exercises and changing transition probabilities
        if model == 'stochastic': 

            action_rewards = np.zeros(self.num_actions)
            next_states = np.zeros(self.num_actions, dtype=int)
            
            action_probabilities = np.zeros(self.num_actions)
            action_probabilities.fill(0.05)
            action_probabilities[environment.actions.index(curr_action)] = 0.85

        elif model == 'deterministic':
             
            action_rewards = np.zeros(self.num_actions)
            next_states = np.zeros(self.num_actions, dtype=int)
            
            action_probabilities = np.zeros(self.num_actions)
            action_probabilities[environment.actions.index(curr_action)] = 1
            
        else:
            raise ValueError('Model was mis-specified!')

        
        # Calculating reward and getting new position for all the actions
        for action_index in range(self.num_actions):
    
            # Determine new position and check whether the agent hits a wall.
            old_position = state
            new_position = state
            if self.actions[action_index] == "n":
                candidate_position = old_position + self.num_cols
                if candidate_position < self.num_fields:
                    new_position = candidate_position
            elif self.actions[action_index] == "e":
                candidate_position = old_position + 1
                if candidate_position % self.num_cols > 0:  # The %-operator denotes "modulo"-division.
                    new_position = candidate_position
            elif self.actions[action_index] == "s":
                candidate_position = old_position - self.num_cols
                if candidate_position >= 0:
                    new_position = candidate_position
            elif self.actions[action_index] == "w":  # "LEFT"
                candidate_position = old_position - 1
                if candidate_position % self.num_cols < self.num_cols - 1:
                    new_position = candidate_position
            else:
                raise ValueError('Action was mis-specified!')

            # Calculate reward
            action_rewards[action_index] = self.rewards[new_position]
            action_rewards[action_index] -= 1
            
            next_states[action_index] = new_position
            
        
        return action_rewards, next_states, action_probabilities
    
```

```python
def play(environment, model):
    
    delta = 1
    theta=1e-10 
    gamma=1.0
    
    policy = np.empty(25, dtype=str)
    policy[:] = "n"
    v = np.zeros(25)

    # Finding Optimal Value Function Using Value Iteration algorithm
    while delta > theta:
        delta = 0

        for state in range(environment.num_fields):
            if state not in [18,23]:
                u = v[state]
                
                sum_array= np.zeros(environment.num_actions)  
                
                # Using Bellman equation
                for action in environment.actions:
                    action_rewards, next_states, action_probabilities = environment.make_step(state, action, model)                    
                    sum_array[environment.actions.index(action)] = np.sum(action_probabilities*(action_rewards + gamma*v[next_states]))
                    
                
                v[state] = np.max(sum_array)
                delta = max(delta, np.abs(u - v[state]))
    
    
    # Extracting Policy from value function only once (since it is not a policy iteration)
    for state in range(environment.num_fields):
        if state not in [18,23]:

            sum_array= np.zeros(environment.num_actions)
            
            for action in environment.actions:
                action_rewards, next_states, action_probabilities = environment.make_step(state, action, model)                    
                sum_array[environment.actions.index(action)] = np.sum(action_probabilities*(action_rewards + gamma*v[next_states]))

            optA_index = np.argmax(sum_array)
            policy[state] = environment.actions[optA_index]
    
    
    return v, policy
```
